# Visualizando dados com Matplotlib


Matplotlib é uma biblioteca de plotagem 2D Python que produz representações de qualidade em uma variedade de formatos impressos e ambientes interativos entre plataformas. Matplotlib pode ser usado em scripts Python, o shell Python e IPython, o notebook jupyter, servidores de aplicativos da web etc

### Matplotlib.Pyplot

Um dos aspectos centrais do Matplotlib é `matplotlib.pyplot`. É a camada de script do Matplotlib.
É uma coleção de funções de estilo de comando que fazem o Matplotlib funcionar como o MATLAB. Cada função `pyplot` faz alguma alteração em uma figura: por exemplo, cria uma figura, cria uma área de plotagem em uma figura, plota algumas linhas em uma área de plotagem, decora a plotagem com rótulos, etc. 

Neste notebook, trabalharemos com a camada de script para aprender como gerar gráficos de linha. Em notebooks futuros, trabalharemos com outras camadas.

Vamos começar importando `Matplotlib` e` Matplotlib.pyplot` da seguinte maneira:

````python
# estamos usando o back-end inline
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
````

Agora, vamos carregar as bibliotecas para análise de dados e o arquivo com o dataset que iremos usar

````python
import numpy as np 
import pandas as pd 
````

````python
dfCanada = pd.read_excel('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx', sheet_name='Canada by Citizenship', skiprows=range(20), skipfooter=2)
dfCanada.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
dfCanada.columns = list(map(str, dfCanada.columns))
dfCanada.set_index('Country', inplace=True)
````

````python
# Antes de começar, vamos dar uma olhada no dataframe
print('infos:', dfCanada.info)
print('data dimensions:', dfCanada.shape)
print(dfCanada.columns)
dfCanada.head(2)
````

### Plotting em _pandas_

Felizmente, o pandas tem uma implementação integrada do Matplotlib que podemos usar. Plotar em _pandas_ é tão simples quanto anexar um método `.plot ()` a uma série ou dataframe.

# Line Pots (Series/Dataframe) <a id="12"></a>


**O que é um gráfico linear e por que usá-lo?**

Um gráfico linear ou gráfico de linha é um tipo de gráfico que exibe informações como uma série de pontos de dados chamados 'marcadores' conectados por segmentos de linha reta. É um tipo básico de gráfico comum em muitos campos.
Use o gráfico de linha quando você tiver um conjunto de dados contínuo. Eles são mais adequados para visualizações de dados baseadas em tendências durante um período de tempo.

**Vamos começar com um estudo de caso:**

Em 2010, o Haiti sofreu um terremoto catastrófico de magnitude 7,0. O terremoto causou devastação generalizada e perda de vidas e cerca de três milhões de pessoas foram afetadas por este desastre natural. Como parte do esforço humanitário do Canadá, o Governo do Canadá intensificou seus esforços ao aceitar refugiados do Haiti. Podemos visualizar rapidamente esse esforço usando um gráfico `Line`:

**Questão:** Faça um gráfico linear da imigração do Haiti usando `df.plot ()`.

Primeiro, extrairemos a série de dados do Haiti.

````python
years = list(map(str, range(1980, 2014))) # Map usa a função str para cada numero no range
haiti = dfCanada.loc['Haiti', years] # passing in years 1980 - 2013 to exclude the 'total' column
haiti.head()
````

````python
haiti.plot()
````

O _pandas_ preenche automaticamente o eixo x com os valores do índice (anos) e o eixo y com os valores da coluna (população). No entanto, observe como os anos não foram exibidos porque são do tipo _string_. Portanto, vamos alterar o tipo dos valores do índice para _integer_ para a plotagem.

Além disso, vamos rotular os eixos xey usando `plt.title ()`, `plt.ylabel ()` e `plt.xlabel ()` da seguinte maneira:

````python
#haiti.index = haiti.index.map(int) # vamos mudar o valor dos indices para inteiro
haiti.plot(kind='line') # tipo de gráfico o tipo padrao eh 'line'
plt.title('Imigrantes do Haiti') # titulo do gráfico
plt.ylabel('Número de Imigrantes') # rotulo do eixo y
plt.xlabel('Anos') # rotulo do eixo x
#plt.show() # precisamos chamar o show() para atualizar o gráfico
````

Podemos notar claramente como o número de imigrantes do Haiti aumentou em 2010, quando o Canadá intensificou seus esforços para aceitar refugiados do Haiti. Vamos anotar esse pico no gráfico usando o método `plt.text ()`.

````python
haiti.plot(kind='line')

plt.title('Imigrantes do Haiti') # titulo do gráfico
plt.ylabel('Número de Imigrantes') # rotulo do eixo y
plt.xlabel('Anos') # rotulo do eixo x

# Anotar o terremoto no gráfico 
plt.text(2000, 6000,'Terremoto de 2010') # indicando a posição no eixo e o texto

plt.show() 
````

Com apenas algumas linhas de código pudemos visualizar rapidamente o pico na imigração!

Nota rápida sobre os valores xey em `plt.text (x, y, label)`:

```
  Como o eixo x (anos) é do tipo 'inteiro', especificamos x como um ano. O eixo y (número de imigrantes) é do tipo 'inteiro', então podemos apenas especificar o valor y = 6000.
```

```python
     plt.text (2000, 6000, 'Terremoto de 2010') # anos armazenados como tipo int
```

```
Se os anos fossem armazenados como tipo 'string', precisaríamos especificar x como a posição do índice do ano. Por exemplo, o 20º índice é o ano 2000, pois é o 20º ano com um ano base de 1980.
```

```python
     plt.text (20, 6000, 'Terremoto de 2010') # anos armazenados como tipo int
```

Podemos facilmente adicionar mais países ao gráfico linear para fazer comparações significativas com a imigração de diferentes países.

Vamos comparar o número de imigrantes da Índia e da China de 1980 a 2013.
Para isso, vamos obter o conjunto de dados para a China e a Índia e exiba o dataframe.

````python
df_CI = dfCanada.loc[['India', 'China'], years]
df_CI.head()
````

````python
# plotando o gráfico

df_CI.plot(kind='line')
````

Lembre-se de que o pandas plota os índices no eixo xe as colunas como linhas individuais no eixo y. Visto que df_CI é um dataframe com o país como índice e anos como colunas, devemos primeiro transpor o dataframe usando o método transpose() para trocar a linha e as colunas.

````python
df_CI = df_CI.transpose()
df_CI.head()
````

````python
df_CI.index = df_CI.index.map(int) 
df_CI.plot(kind='line')

plt.title('Imigrantes de China e India') # titulo do gráfico
plt.ylabel('Número de Imigrantes') # rotulo do eixo y
plt.xlabel('Anos') # rotulo do eixo x

plt.show()
````
