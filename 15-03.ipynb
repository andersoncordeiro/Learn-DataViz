## pie

```python
df_continents = dfCanada.groupby('Continent', axis=0).sum()
df_continents.head()
```

```python
df_continents['Total'].plot(kind='pie', shadow=True,  figsize=(8,5), autopct='%1.1f%%')
plt.title('Imigrantes por continente')
plt.show()
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0, 0.1, 0, 0, 0, 0]

df_continents['Total'].plot(kind='pie', 
                            shadow=False,
                            labels=None,
                            startangle=90, 
                            figsize=(15,10), 
                            autopct='%1.1f%%',
                            colors=colors_list,
                            explode=explode_list
                           )

plt.legend(labels=df_continents.index, loc='upper left')
plt.ylabel(None)
```

### Scatter

```python
df_total = pd.DataFrame(dfCanada[years].sum(axis=0))
```

```python
df_total.index = map(int, df_total.index)
```

```python
df_total.reset_index(inplace=True)
```

```python
df_total.head()
```

```python
df_total.columns = ['year', 'total']
```

```python
df_total.head()
```

```python
df_total.plot(kind='scatter', x='year', y='total')
plt.title('Total de Imigrantes para o Canada')
plt.show()
```
